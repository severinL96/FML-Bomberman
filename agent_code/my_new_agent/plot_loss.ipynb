```python
import pickle
import matplotlib.pyplot as plt
import numpy as np


LOSS = open("saved_models/Dense_2000_epochs_more_random/loss.txt", 'r',)

losses = []
for loss in LOSS.readlines():
    loss = loss.rstrip('\n')
    loss = float(loss)
    losses.append(loss)

epoch_count = range(1,len(losses)+1)


coefficients = np.polyfit(epoch_count,losses,4)
poly = np.poly1d(coefficients)
new_y = poly(epoch_count)


plt.xlabel('#Rounds', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.yscale('log')
plt.plot(epoch_count,losses)
plt.plot(epoch_count, new_y)
plt.grid(True, which="both")
plt.title('loss Dense - Decreasing randomness')
plt.Figure()
```
